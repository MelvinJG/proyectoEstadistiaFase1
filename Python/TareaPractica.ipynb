### <center> Histograma de frecuencia (Análisis Bivariado)  </center>

```python
import pandas as pd
import matplotlib.pyplot as plt

myData = pd.read_excel('MUESTRAPOBLACIONAL.xlsx') 
df = pd.DataFrame(myData)

## Punteo de Matemática de personas que SI se prepararon
MathYes = df[(df['Preparación'] == 'si')]
print(f"Personas que SI se prepararon: {MathYes['Matemática'].count()}")

## Punteo de Matemática de personas que NO se prepararon
MathNo = df[(df['Preparación'] == 'no')]
print(f"Personas que NO se prepararon: {MathNo['Matemática'].count()}")

## Grafica SI se prepararon
y= MathYes["Matemática"]
plt.hist(y, bins = 10, color = "blue", rwidth=0.9, orientation="horizontal")
plt.title("Histograma de frecuencia variable punteo de Matemática personas que SI se prepararon")
plt.xlabel("Punteos")
plt.ylabel("Frecuencia")
plt.show()
```

```python
## Grafica No se prepararon
y= MathNo["Matemática"]
plt.hist(y, bins = 10, color = "blue", rwidth=0.9, orientation="horizontal")
plt.title("Histograma de frecuencia variable punteo de Matemática personas que NO se prepararon")
plt.xlabel("Punteos")
plt.ylabel("Frecuencia")
plt.show()
```

```python
## Resumen grafica
names = ['Se preparo', 'No se preparo']
values = [MathYes['Matemática'].sum(), MathNo['Matemática'].sum()]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(131)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Resumen del punteo de Matemática en función de la preparación')
plt.show()
```

### <center> Histograma de frecuencia (Análisis Bivariado)  </center>

```python
import pandas as pd
import matplotlib.pyplot as plt

myData = pd.read_excel('MUESTRAPOBLACIONAL.xlsx') 
df = pd.DataFrame(myData)

Extranjero = df[(df['Etnia'] == 'Extranjero')]
print(f"Personas extranjeras: {Extranjero['Lectura'].count()}")

xinca = df[(df['Etnia'] == 'xinca')]
print(f"Personas xincas: {xinca['Lectura'].count()}")

garifuna = df[(df['Etnia'] == 'garifuna')]
print(f"Personas garifunas: {garifuna['Lectura'].count()}")

ladino = df[(df['Etnia'] == 'ladino')]
print(f"Personas ladinas: {ladino['Lectura'].count()}")

maya = df[(df['Etnia'] == 'maya')]
print(f"Personas mayas: {maya['Lectura'].count()}")
```

```python
## Resumen grafica
names = ['Ext', 'Xin', 'Gari','Ladi','May']
values = [Extranjero['Lectura'].sum(), xinca['Lectura'].sum(), garifuna['Lectura'].sum(), ladino['Lectura'].sum(), maya['Lectura'].sum()]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(131)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Resumen del punteo de Lectura en función de su Etnia')
plt.show()
print('''Ext = Extranjeros
Xin = Xincas
Gari = Garifunas
Ladi = Ladinas
May = Mayas
''')
```
